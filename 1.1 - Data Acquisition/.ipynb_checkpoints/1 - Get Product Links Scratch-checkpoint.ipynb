```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import pickle
import time, os
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
driver = webdriver.Chrome(chromedriver)
```

```python
base_url = 'https://www.vino.com/en/s/?nav=true&categoria=vini&ultimaScelta=abbinamenti'
driver.get(base_url)
response = requests.get(base_url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
response.status_code
```

```python
for i in range(20):
    # scroll
    driver.execute_script(
        "window.scrollTo(0, document.documentElement.scrollHeight);" #Alternatively, document.body.scrollHeight
    )
    ##### takes 4 scrolls to get down to 
    time.sleep(1)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
all_base_content = soup.find_all('div', class_ = "col-xs-12 col-md-4 text-center ias-item")
base_page_links = []

for  div in all_base_content:
    for div1 in div.find_all('div', class_ = "col-xs-9 search-result-bottle-grid-content"):
        ## could consider creating a list in the above cell for this one called like all_base_content_grids
        for links in div1.find_all('a'):
            link = links.get("href")
            base_page_links.append(link)
    # page will not load if you don't wait for it
```

```python
base_page_links
```

```python
print(len(base_page_links))
```

```python
with open('base_page_links.pkl', 'wb') as f: 
    pickle.dump(base_page_links, f)
```
