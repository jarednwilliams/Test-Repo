```python
import ast
```

```python
def CleanLinks(link):
    with open(link) as f:
        links = f.read()
    link_list = list(set(ast.literal_eval(links)))
    return link_list
```

```python
all_product_links = []
```

```python
product_links_1 = CleanLinks('product_links_1.txt')
all_product_links.extend(product_links_1)
```

```python
product_links_2 = CleanLinks('product_links_2.txt')
all_product_links.extend(product_links_2)
```
