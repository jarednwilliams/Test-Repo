```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import pickle
import time, os
import numpy as np
from numpy import random
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

#### Creating List of Base Page Links from which the Products come from

```python
base_page_link_list = ['https://www.wine.com/list/wine/7155/?showOutOfStock=false']
for i in range(2,661):
    base_page_link_list.append('https://www.wine.com/list/wine/7155/'+str(i)+'?showOutOfStock=false')
base_page_link_list[:5]
```

#### Initalize Web Driver from Selenium

```python
chromedrive = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedrive
driver = webdriver.Chrome(chromedrive)
```

```python
def base_url(url):
    driver.get(url)
    response = requests.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    all_base_content = soup.find_all('li', class_ = "prodItem")
    bottle_page_links = []
    
    for  div in all_base_content:
        for div1 in div.find_all('div', class_ = "prodItemInfo"):
            for links in div1.find_all('a'):
                link = links.get("href")
                bottle_page_links.append(link)
                
    for i in bottle_page_links:
        if i == "#":
            bottle_page_links.remove(i)
    
    updated_bottle_page_links = []
    for i in bottle_page_links:
        updated_bottle_page_links.append('https://www.wine.com' + i)
    
    return updated_bottle_page_links
```

### Scrape First 2500 Products

```python
#### Setting Sleeptime
sleeptime = round(random.uniform(8, 13))
sleeptime
```

```python
#### Scraping to get Empty List need to change 5 to 100 later...
product_list = []
counter = 0
for page in base_page_link_list[:100]:
    product_list.extend(base_url(page))
    counter += 1
    if counter%100 == 0:
        with open('product_links.txt', 'w') as f: 
            f.write(str(product_list))
    time.sleep(sleeptime)
```
